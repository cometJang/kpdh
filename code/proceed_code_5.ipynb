```python
import pandas as pd

df=pd.read_csv('../data/raw/museum/csv/국립중앙박물관 관람객수_daily_2015_2025.csv')
print(df.head())
```

```python
df.dtypes
```

```python
# visitors: 결측치를 허용하는 정수형으로 변환
df["visitors"] = pd.to_numeric(df["visitors"], errors="coerce").astype("int64")

# date: 시계열 데이터 타입으로 변환
df["date"] = pd.to_datetime(df["date"], errors="coerce")

# 결과 확인
print(df.dtypes)
print(df.head())

display(df.head())
```

```python
# # 여러 컬럼을 한꺼번에 처리할 때 (예시)
# cols_to_fix = ["visitors",'korean', "foreigner"]
# df[cols_to_fix] = df[cols_to_fix].apply(pd.to_numeric, errors="coerce").astype("int64")
# df.dtypes
```

```python
df=df.drop(columns=['korean','foreigner'])
```

```python
# 주말(토, 일)이면 True, 아니면 False
df["is_weekend"] = df["date"].dt.dayofweek >= 5
```

```python
display(df.head())
```

```python
# 1. 방문자가 0인 데이터만 추출
zero_visitors_df = df.loc[df["visitors"] == 0]
# zero_visitors_df = df.loc[df["visitors"] == 0, ["date", "day_of_week"]]

# 2. 상세 내역 출력
print("### 방문객이 0명인 날짜 상세 리스트 ###")
display(zero_visitors_df)

to_csv_path = "../data/processed/zero_visitors.csv"
zero_visitors_df.to_csv(to_csv_path, index=False)

# 3. 전체 개수 출력
# print(f"\n총 0명이었던 일수: {len(zero_visitors_df)}일")
```
