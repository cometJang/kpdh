```python
import pandas as pd

# # 엑셀 파일 읽기
# # file_path = r'C:\Users\Comet\Desktop\kpdh\data\raw\museum\csv\국립박물관문화재단_박물관상품 인기관심도.xlsx'
# file_path = '../data/raw/museum/csv/국립박물관문화재단_박물관상품 인기관심도.xlsx'

# # 각 시트 읽기
# df_2023 = pd.read_excel(file_path, sheet_name='2023')
# df_2024 = pd.read_excel(file_path, sheet_name='2024')
# df_2025 = pd.read_excel(file_path, sheet_name='2025')

# 1. 파일 경로를 문자열로 지정
file_path = '../data/raw/museum/csv/국립박물관문화재단_박물관상품 인기관심도.xlsx'

# 2. ExcelFile 객체 생성 (파일을 한 번만 로드해서 효율적임)
excel_data = pd.ExcelFile(file_path)

# 3. 각 시트 읽기
df_2023 = excel_data.parse('2023')
df_2024 = excel_data.parse('2024')
df_2025 = excel_data.parse('2025')

# 결과 확인
display(df_2023.head())
display(df_2024.head())
display(df_2025.head())
```

```python
# 연도 컬럼 추가
df_2023['year'] = 2023
df_2024['year'] = 2024
df_2025['year'] = 2025

# 컬럼 순서 통일 (2024가 다름)
df_2023 = df_2023[['year', 'quarter', 'rank', 'category', 'product_name']]
df_2024 = df_2024[['year', 'quarter', 'rank', 'category', 'product_name']]
df_2025 = df_2025[['year', 'quarter', 'rank', 'category', 'product_name']]

# 데이터프레임 병합
df = pd.concat([df_2023, df_2024, df_2025], ignore_index=True)

display(df)
```

```python
# 정렬 (연도 → 분기 → 순위)
df = df.sort_values(['year', 'quarter', 'rank']).reset_index(drop=True)
display(df)
df.dtypes
```

```python
# 결과 확인
print(f"전체 shape: {df.shape}")
print(f"\n연도별 데이터 수:")
print(df['year'].value_counts().sort_index())
print(f"\n최종 데이터:")
display(df.head(10))
display(df.tail(10))
```
